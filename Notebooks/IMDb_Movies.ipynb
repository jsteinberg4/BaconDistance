<center><h1> Several columns contain data which will be irrelevant to the Bacon Distance computation </h1></center>

<ul>
    <li>
        Any column containing information about scores, voting, or reviews. This is too generic and doesn't create a concrete connection from one actor/movie to another, so we can drop it.
    </li>
    <li>
        Any column containing budget or income information. This is also generic and doesn't provide a usable link between films.
    </li>
    <li>
        The description column. While it's possible this could contain usable information (i.e an explicit mention of Kevin Bacon), it would require too much processing overhead in addition to a graph search to be useful. For now we'll drop the column, but it could be added back later as an additional feature.
    </li>
</ul>

```python
# Lists of all the columns we'll be dropping
score_cols = ["avg_vote", "votes", "metascore", "reviews_from_critics", "reviews_from_users"]
money_cols = ["budget", "usa_gross_income", "worlwide_gross_income"]
description_col = ["description"]
```

```python
# First drop scoring and review information...
drop_scores = movies.drop(score_cols, axis=1)
drop_scores.head()
```

```python
# Next drop the monetary info...
drop_money = drop_scores.drop(money_cols, axis=1)
drop_money.head()
```

```python
# Then drop the description...
drop_desc = drop_money.drop(description_col, axis=1)
drop_desc.head()
```
